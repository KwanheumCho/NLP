```python
def tfidf_extractor(corpus,):
    # returns a tf-idf based DTM
    vectorizer = TfidfVectorizer(min_df=1, ngram_range=(1,1))
    features = vectorizer.fit_transform(corpus)
    return vectorizer, features
```

# 회귀 모델 학습 (Lasso Regression) 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression(C=0.1, penalty='l1', solver='saga') # Lasso regression

# C = Inverse of regularization strength, 즉 C 값이 작을수록 penalty를 많이 준다는 것입니다.
# penalty를 많이 준다는 뜻은 L1 같은 경우는 feature의 수를 그만큼 많이 줄인다는 뜻이고
# L2인 경우는 weight 값을 더 0에 가깝게 한다는 뜻입니다.

lr.fit(features, label_train) # 학습
```

# 회귀 모델로 긍정단어 50개, 부정단어 50개 뽑기  

```python
# Get coefficients of the model
coefficients = lr.coef_.tolist()

sorted_coefficients = sorted(enumerate(coefficients[0]), key=lambda x:x[1], reverse=True)
# 학습에 사용된 각 단어마다의 coefficient (즉 weight) 값이 존재
# coefficient값이 큰 순으로 정렬 'reverse=True'

# print top 50 positive words
print("Top 50 긍정 단어")
for word, coef in sorted_coefficients[:50]:
    print('{0:} ({1:.3f})'.format(vocablist[word], coef))
print("===================")

# print top 50 negative words
print("Top 50 부정단어")
for word, coef in sorted_coefficients[-50:]:
    print('{0:} ({1:.3f})'.format(vocablist[word], coef))
```

```python

def filter(s):  # 한글만 남김
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    result = hangul.sub("",s)
    return result 
    
def preprocessing_data (test_data, postprocessor): # 불용어 제거 및 noun, adj, adverb만 남기기 
    
    # 한글만 남김
    for i in range(len(test_data)):
        test_data[i] = filter(test_data[i])
    
    # 불용어 제거 및 형태소별 정리  # adj, adverb, noun만 남기기 
    test_pos = []
    for lines in test_data:
        tmp = postprocessor.pos(lines)
        test_pos.append(tmp)
    
    
    # string 형태로 저장
    test_string = []
    for lines in test_pos: 
        tmp1 = ""
        i=0
        for words in lines: 
            if i==0:
                tmp1 = words[0]
            else: 
                tmp1 = tmp1 + " " + words[0]
            i+=1
        test_string.append(tmp1)
    
    return test_string

def preprocessing_data_list (test_data, postprocessor): # 불용어 제거 및 noun, adj, adverb만 남기기 
    
    # 한글만 남김
    for i in range(len(test_data)):
        test_data[i] = filter(test_data[i])
     
    # 불용어 제거 및 형태소별 정리 # adj, adverb, noun만 남기기 
    test_pos = []
    for lines in test_data:
        tmp = postprocessor.pos(lines)
        test_pos.append(tmp)
    
    
    
    # 리스트 형태로 저장
    test_list = []
    for lines in test_pos: 
        tmp1 = []
        for words in lines: 
            tmp1.append(words[0])
        test_list.append(tmp1)
    
    return test_list

def preprocessing_label(test_label): # 4점 이상이면 긍정(1), 3점 이하면 부정(0)
    result = []
    for stars in test_label:
        if stars >3.0:
            result.append(1)
        else:
            result.append(0)
    return result
```

```python
# 댓글마다 주제어 선정 

import operator
from collections import Counter

# 양 가격 - 가성비 
# 친절 기분 직원 사장- 서비스
# 맛 식사 - 음식
# 조용 - 분위기

def decide_category(test_list, word_vectors, model):
    similarity = []
    for sen in test_list:
        tmp_sen = []
        for word in sen:
            tmp = {}
            if word in word_vectors.vocab:
                tmp['분위기'] = model.similarity(word, '분위기')
                tmp['서비스'] = model.similarity(word, '서비스')
                tmp['음식'] = model.similarity(word, '음식')
                tmp['양'] = model.similarity(word, '양')
                tmp['가격'] = model.similarity(word, '가격')
                tmp['친절'] = model.similarity(word, '친절')
                tmp['기분'] = model.similarity(word, '기분')
                tmp['직원'] = model.similarity(word, '직원')
                tmp['맛'] = model.similarity(word, '맛')
                tmp['식사'] = model.similarity(word, '식사')
                tmp['조용'] = model.similarity(word, '조용')
                tmp['사장'] = model.similarity(word, '사장')
                tmp_sen.append(max(tmp.items(), key=operator.itemgetter(1))[0])
            else: 
                tmp_sen.append('음식')
        counter = Counter(tmp_sen)
        maxx = max(tmp_sen, key=counter.get)
        
        if maxx =="양" or maxx =="가격":
            similarity.append("가성비")
        elif maxx =="친절" or maxx == "기분" or maxx == "직원" or maxx == "사장":
            similarity.append("서비스")
        elif maxx =="맛" or maxx == "식사":
            similarity.append("음식")
        elif maxx == "조용":
            similarity.append("분위기")
        else:
            similarity.append(maxx)
        
    return similarity
            
```
